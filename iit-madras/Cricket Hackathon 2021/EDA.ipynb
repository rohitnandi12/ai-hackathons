<tr>
<td><code>start_date</code></td>
<td><strong>string (required)</strong> The start date of the match took place, in the format <code>YYYY/MM/DD</code>. Matches with multiple days will only show the first.</td>

<tr>
<td><code>venue</code></td>
<td><strong>string</strong> The venue in which the game took place, such as <strong>Old Trafford</strong>. If not provided this will be blank.</td>

<td><code>batting_team</code></td>
<td><strong>string (required)</strong> The name of the team which was batting for the delivery.</td>

<td><code>bowling_team</code></td>
<td><strong>string (required)</strong> The name of the team which was bowling for the delivery.</td>
</tr>
<tr>
<td><code>striker</code></td>
<td><strong>string (required)</strong> The name of the batter who faced the delivery.</td>
</tr>
<tr>
<td><code>non_striker</code></td>
<td><strong>string (required)</strong> The name of the non-striker during the delivery.</td>
</tr>
<tr>
<td><code>bowler</code></td>
<td><strong>string (required)</strong> The name of the bowler during the delivery.</td>
</tr>
<tr>
<td><code>runs_off_bat</code></td>
<td><strong>integer (required)</strong> The total number of runs scored by the batter off the ball. If the batter failed to score this will show <strong>0</strong></td>
</tr>
<tr>
<td><code>extras</code></td>
<td><strong>integer (required)</strong> The total number of runs conceded via extras off the ball. If no extras were conceded this will show <strong>0</strong>.</td>
</tr>
<tr>
<td><code>wides</code></td>
<td><strong>integer</strong> The number of wides conceded on the delivery. If no wides were conceded this will be blank.</td>
</tr>
<tr>
<td><code>noballs</code></td>
<td><strong>integer</strong> The number of no-balls conceded on the delivery. If no no-balls were conceded this will be blank.</td>
</tr>

<td><code>byes</code></td>
<td><strong>integer</strong> The number of byes conceded on the delivery. If no byes were conceded this will be blank.</td>
</tr>

<td><code>legbyes</code></td>
<td><strong>integer</strong> The number of legbyes conceded on the delivery. If no leg-byes were conceded this will be blank.</td>
</tr>
<tr>
<td><code>penalty</code></td>
<td><strong>integer</strong> The number of penalty runs conceded on the delivery. If no penalty runs were conceded this will be blank.</td>
</tr>
<tr>
<td><code>wicket_type</code></td>
<td><strong>string</strong> If a wicket fell on this delivery then this will be the kind of dismissal that took place. This will be one of <strong>bowled</strong>, <strong>caught</strong>, <strong>caught and bowled</strong>, <strong>lbw</strong>, <strong>stumped</strong>, <strong>run out</strong>, <strong>retired hurt</strong>, <strong>hit wicket</strong>, <strong>obstructing the field</strong>, <strong>hit the ball twice</strong>, <strong>handled the ball</strong>, or <strong>timed out</strong>. If no dismissals took place this will be an empty string.</td>
</tr>
<tr>
<td><code>player_dismissed</code></td>
<td><strong>string</strong> If a wicket fell on this delivery then this will be the name of the dismissed player. If no dismissals took place this will be an empty string.</td>
</tr>
<tr>
<td><code>other_wicket_type</code></td>
<td><strong>string</strong> If a second wicket occurs on the delivery (such as due to a player retiring hurt when another wicket falls) this will be the type of the dismissal, one of <strong>bowled</strong>, <strong>caught</strong>, <strong>caught and bowled</strong>, <strong>lbw</strong>, <strong>stumped</strong>, <strong>run out</strong>, <strong>retired hurt</strong>, <strong>hit wicket</strong>, <strong>obstructing the field</strong>, <strong>hit the ball twice</strong>, <strong>handled the ball</strong>, or <strong>timed out</strong>. If no dismissals took place this will be an empty string.</td>
</tr>
<tr>
<td><code>other_player_dismissed</code></td>
<td><strong>string</strong> If a second wicket fell on this delivery then this will be the name of the second dismissed player. If no dismissals took place this will be an empty string.</td>
</tr>
